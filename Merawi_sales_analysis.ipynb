## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

# Load Data Files

```python
# StoreSales all in-store transactions
store_sales = pd.read_csv('StoreSales.csv')
store_sales.info()
```

```python
# StoreDetail store locations, territories, and regions
store_detail = pd.read_csv('StoreDetail.csv')
store_detail.info()
```

```python
# Products product catalog with category references
products = pd.read_csv('Products.csv')
products.info()
```

```python
# ProductCategories category and subcategory names
product_categories = pd.read_csv('ProductCategories.csv')
product_categories.info()
```

```python
# customer_list rewards members 
customer_list = pd.read_csv('customer_list.csv', sep='|')
customer_list.info()
```

## Data Preparation

```python
# Convert Transaction Date to datetime
store_sales['Transaction Date'] = pd.to_datetime(store_sales['Transaction Date'])

# Add Month and Year columns for time based grouping
store_sales['Month'] = store_sales['Transaction Date'].dt.to_period('M')
store_sales['Year'] = store_sales['Transaction Date'].dt.year

# Filter stores for Massachusetts and Maryland territories
territory_stores = store_detail[store_detail['State'].isin(['Massachusetts', 'Maryland'])].copy()
territory_ids = territory_stores['Store ID'].tolist()

# Filter sales to my two territories
territory_sales = store_sales[store_sales['Store ID'].isin(territory_ids)].copy()

print(f'Total territory transactions: {len(territory_sales):,}')
print(f'Date range: {territory_sales["Transaction Date"].min().date()} to {territory_sales["Transaction Date"].max().date()}')
print(f'Stores in territory: {territory_sales["Store ID"].nunique()}')
```

## Core Marketing Analysis

### Question 1 Territory Managers, Store IDs, and Cities

Who are the territory managers for each assigned territory? What are the store IDs and cities for each store?

```python
# Massachusetts stores
ma_stores = territory_stores[territory_stores['State'] == 'Massachusetts'][['Store ID', 'Store Location', 'Territory Manager', 'Region', 'Region Director']]
print('MASSACHUSETTS')
print(ma_stores.to_string(index=False))
```

```python
# Maryland stores
md_stores = territory_stores[territory_stores['State'] == 'Maryland'][['Store ID', 'Store Location', 'Territory Manager', 'Region', 'Region Director']]
print('MARYLAND')
print(md_stores.to_string(index=False))
```

### Question 2 Monthly Total Revenue by Territory

What is the monthly total revenue for in-store sales in each territory over the full data period?

```python
# Merge sales with store state info
sales_with_state = territory_sales.merge(
    territory_stores[['Store ID', 'State']], on='Store ID'
)

# Group by Month and State
monthly_revenue = sales_with_state.groupby(['Month', 'State'])['Sale Amount'].sum().reset_index()
monthly_revenue.columns = ['Month', 'State', 'Total Revenue']

# Pivot for readability
monthly_pivot = monthly_revenue.pivot(index='Month', columns='State', values='Total Revenue').reset_index()
monthly_pivot.columns.name = None
monthly_pivot['Month'] = monthly_pivot['Month'].astype(str)
print(monthly_pivot.to_string(index=False))
```

```python
# CHART 1 Monthly Revenue Over Time by Territory 
ma_monthly = monthly_revenue[monthly_revenue['State'] == 'Massachusetts'].copy()
md_monthly = monthly_revenue[monthly_revenue['State'] == 'Maryland'].copy()

ma_monthly['Month_str'] = ma_monthly['Month'].astype(str)
md_monthly['Month_str'] = md_monthly['Month'].astype(str)

fig, ax = plt.subplots(figsize=(14, 6))

ax.plot(ma_monthly['Month_str'], ma_monthly['Total Revenue'], label='Massachusetts', color='steelblue', linewidth=2, marker='o', markersize=3)
ax.plot(md_monthly['Month_str'], md_monthly['Total Revenue'], label='Maryland', color='darkorange', linewidth=2, marker='s', markersize=3)

ax.set_title('Monthly In-Store Revenue by Territory (Jan 2022 – Dec 2025)', fontsize=14, fontweight='bold')
ax.set_xlabel('Month', fontsize=11)
ax.set_ylabel('Total Revenue ($)', fontsize=11)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))

# Show every 6th label to avoid crowding
tick_positions = range(0, len(ma_monthly), 6)
ax.set_xticks([ma_monthly['Month_str'].iloc[i] for i in tick_positions])
plt.xticks(rotation=45, ha='right')

ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print('\nChart type: Line chart — best for showing revenue trends over time')
```

### Result
- Maryland consistently generates higher monthly revenue than Massachusetts throughout the data period.
- Maryland revenue ranges from 190K in early 2022 up to 360K by late 2025, showing strong growth over the full period. Massachusetts grew from 69K to over 277K an even stronger relative growth rate.

### Question 3 Store Performance Rankings

How would you rank the sales performance of each store in each territory? Which are the top performers?

```python
# Total revenue per store merged with store info
store_revenue = sales_with_state.groupby(['Store ID', 'State'])['Sale Amount'].agg(
    Total_Revenue='sum',
    Num_Transactions='count',
    Avg_Transaction='mean'
).reset_index()

store_revenue = store_revenue.merge(territory_stores[['Store ID', 'Store Location']], on='Store ID')
store_revenue['Rank'] = store_revenue.groupby('State')['Total_Revenue'].rank(ascending=False).astype(int)
store_revenue = store_revenue.sort_values(['State', 'Rank'])

print(' MASSACHUSETTS — Store Rankings')
ma_rank = store_revenue[store_revenue['State'] == 'Massachusetts'][['Rank', 'Store ID', 'Store Location', 'Total_Revenue', 'Num_Transactions', 'Avg_Transaction']]
ma_rank['Total_Revenue'] = ma_rank['Total_Revenue'].map('${:,.2f}'.format)
ma_rank['Avg_Transaction'] = ma_rank['Avg_Transaction'].map('${:.2f}'.format)
print(ma_rank.to_string(index=False))

print('\n MARYLAND — Store Rankings')
md_rank = store_revenue[store_revenue['State'] == 'Maryland'][['Rank', 'Store ID', 'Store Location', 'Total_Revenue', 'Num_Transactions', 'Avg_Transaction']]
md_rank['Total_Revenue'] = md_rank['Total_Revenue'].map('${:,.2f}'.format)
md_rank['Avg_Transaction'] = md_rank['Avg_Transaction'].map('${:.2f}'.format)
print(md_rank.to_string(index=False))
```

```python
# CHART 2 Store Revenue Comparison 
store_rev_raw = sales_with_state.groupby(['Store ID', 'State'])['Sale Amount'].sum().reset_index()
store_rev_raw = store_rev_raw.merge(territory_stores[['Store ID', 'Store Location']], on='Store ID')

ma_bars = store_rev_raw[store_rev_raw['State'] == 'Massachusetts'].sort_values('Sale Amount', ascending=False)
md_bars = store_rev_raw[store_rev_raw['State'] == 'Maryland'].sort_values('Sale Amount', ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Massachusetts
axes[0].barh(ma_bars['Store Location'], ma_bars['Sale Amount'], color='steelblue')
axes[0].set_title('Massachusetts — Total Revenue by Store', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Total Revenue ($)')
axes[0].xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x/1e6:.1f}M'))
axes[0].invert_yaxis()

# Maryland
axes[1].barh(md_bars['Store Location'], md_bars['Sale Amount'], color='darkorange')
axes[1].set_title('Maryland — Total Revenue by Store', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Total Revenue ($)')
axes[1].xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x/1e6:.1f}M'))
axes[1].invert_yaxis()

plt.suptitle('Store Performance Rankings by Territory', fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()

print('Chart type: Horizontal bar chart — best for comparing store level performance across a category')
```

### Results
- Maryland top performer - North Harford - the highest revenue store in either territory at 8.7M, an outlier compared to all other stores.
- Massachusetts top performer - Worcester - leads MA with 602K.

### Question 4 Top Customers by Territory

Comparing the RewardsID from sales data with cust_id from the customer list, who were the top customers in each territory?

```python
# Only keep rewards transactions 
rewards_sales = sales_with_state[sales_with_state['RewardsID'].notna()].copy()
rewards_sales['RewardsID'] = rewards_sales['RewardsID'].astype(int)
customer_list['cust_id'] = customer_list['cust_id'].astype(int)

# Total spend per RewardsID per State
top_cust = rewards_sales.groupby(['RewardsID', 'State'])['Sale Amount'].agg(
    Total_Spend='sum',
    Transactions='count'
).reset_index()

# Merge with customer names
top_cust = top_cust.merge(customer_list[['cust_id', 'name']], left_on='RewardsID', right_on='cust_id', how='left')

print('TOP 5 CUSTOMERS — Massachusetts')
ma_cust = top_cust[top_cust['State'] == 'Massachusetts'].sort_values('Total_Spend', ascending=False).head(5)
ma_cust['Total_Spend'] = ma_cust['Total_Spend'].map('${:,.2f}'.format)
print(ma_cust[['name', 'RewardsID', 'Total_Spend', 'Transactions']].to_string(index=False))

print('\n TOP 5 CUSTOMERS — Maryland')
md_cust = top_cust[top_cust['State'] == 'Maryland'].sort_values('Total_Spend', ascending=False).head(5)
md_cust['Total_Spend'] = md_cust['Total_Spend'].map('${:,.2f}'.format)
print(md_cust[['name', 'RewardsID', 'Total_Spend', 'Transactions']].to_string(index=False))
```

### Result
- Massachusetts top customer is Beverly Crusher who spent 4,758 across 14 transactions.
- Maryland top customer is Cole Brown who spent 7,602 across 26 transactions.
- Maryland top customers spend significantly more than Massachusetts top customers.
- I matched RewardsID from StoreSales with cust_id from customer_list to identify these customers.

### Question 5 Transactions and Revenue by Product Category

What is the number of transactions per month by product category in each territory? What is total revenue per month by category?

```python
# Merge products with categories
prod_with_cats = products.merge(product_categories, on=['CategoryID', 'SubcategoryID'])

# Merge territory sales with product category info
territory_full = sales_with_state.merge(prod_with_cats[['Prod Num', 'Product', 'Category', 'Subcategory']], on='Prod Num', how='left')

# Category totals per territory
cat_totals = territory_full.groupby(['State', 'Category']).agg(
    Total_Revenue=('Sale Amount', 'sum'),
    Transactions=('Sale Amount', 'count')
).reset_index().sort_values(['State', 'Total_Revenue'], ascending=[True, False])

print('MASSACHUSETTS — Revenue & Transactions by Category')
ma_cat = cat_totals[cat_totals['State'] == 'Massachusetts'].copy()
ma_cat['Total_Revenue'] = ma_cat['Total_Revenue'].map('${:,.2f}'.format)
print(ma_cat[['Category', 'Total_Revenue', 'Transactions']].to_string(index=False))

print('\n MARYLAND — Revenue & Transactions by Category ')
md_cat = cat_totals[cat_totals['State'] == 'Maryland'].copy()
md_cat['Total_Revenue'] = md_cat['Total_Revenue'].map('${:,.2f}'.format)
print(md_cat[['Category', 'Total_Revenue', 'Transactions']].to_string(index=False))
```

```python
# Monthly transactions and revenue by category per territory
cat_monthly = territory_full.groupby(['Month', 'State', 'Category']).agg(
    Transactions=('Sale Amount', 'count'),
    Revenue=('Sale Amount', 'sum')
).reset_index()

cat_monthly['Revenue'] = cat_monthly['Revenue'].map('${:,.2f}'.format)
cat_monthly['Month'] = cat_monthly['Month'].astype(str)

print('MASSACHUSETTS — Monthly Transactions & Revenue by Category')
ma_monthly = cat_monthly[cat_monthly['State'] == 'Massachusetts'].sort_values(
    ['Month', 'Category']
).reset_index(drop=True)
print(ma_monthly[['Month', 'Category', 'Transactions', 'Revenue']].to_string(index=False))

print('\n MARYLAND — Monthly Transactions & Revenue by Category')
md_monthly = cat_monthly[cat_monthly['State'] == 'Maryland'].sort_values(
    ['Month', 'Category']
).reset_index(drop=True)
print(md_monthly[['Month', 'Category', 'Transactions', 'Revenue']].to_string(index=False))
```

```python
# CHART 3 Revenue by Category per Territory
cat_rev_raw = territory_full.groupby(['State', 'Category'])['Sale Amount'].sum().reset_index()

fig, axes = plt.subplots(1, 2, figsize=(16, 6))
colors = ['#2196F3', '#F44336', '#4CAF50', '#FF9800', '#9C27B0', '#00BCD4']

for ax, state in zip(axes, ['Massachusetts', 'Maryland']):
    data = cat_rev_raw[cat_rev_raw['State'] == state].sort_values('Sale Amount', ascending=False).reset_index(drop=True)
    x = range(len(data))
    bars = ax.bar(x, data['Sale Amount'], color=colors[:len(data)])
    ax.set_title(f'{state} — Revenue by Product Category', fontsize=12, fontweight='bold')
    ax.set_ylabel('Total Revenue ($)')
    ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f'${v/1e6:.1f}M'))
    ax.set_xticks(list(x))
    ax.set_xticklabels(data['Category'], rotation=30, ha='right', fontsize=9)

    # Add value labels on top of bars
    for bar in bars:
        h = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, h * 1.01,
                f'${h/1e6:.2f}M', ha='center', va='bottom', fontsize=8)

plt.suptitle('Total Revenue by Product Category — Massachusetts vs Maryland',
             fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print('Chart type Bar chart — best for comparing revenue totals across product categories')
```

**Most Popular Products:**
- Technology and Accessories is the most popular category by revenue in both territories
- Laptops are the single biggest subcategory driving tech revenue
- Textbooks rank second in both territories expected for a student bookstore
- Stationery has the MOST transactions in both territories but the least revenue customers buy many cheap items like pens and notebooks

**Growth Opportunities:**
- Stationery and Supplies It has the most customers visiting but lowest revenue. Bundling with higher value items could increase revenue per visit
- Books General is very weak in Massachusetts - running book promotions could help grow this category
- Apparel has strong customer traffic but low revenue per sale - better quality branded merchandise could boost this category

### Question 6 Marketing Recommendation for Next Quarter

Based on the full analysis where should the marketing team focus attention in the next quarter?

### 1. Focus Marketing On Top Customers
Our analysis shows that top customers in Maryland spend nearly double compared to top customers in Massachusetts. Cole Brown in Maryland spent 7,602 while Beverly Crusher in Massachusetts spent $4,758. The marketing team should identify and reward these high value customers with special offers and loyalty benefits to keep them coming back and encourage them to spend more.

### 2. Bring More Customers Into Massachusetts Stores
Maryland makes more total revenue but Massachusetts customers actually spend more money per visit. The problem is not how much they spend it is that they are not coming in often enough. Better local marketing in lower performing stores like Lowell, Framingham, and Somerville could help.

### 3. Learn From Store 736 in North Harford Maryland
This one store makes more revenue than all 18 Massachusetts stores combined. It is a huge outlier. Finding out why it is so successful whether it is near a university has more customers or sells differently could help improve other stores.

### 4. Increase Revenue From Apparel and Art Supplies
These two categories have many customers coming in but each purchase is very cheap. Bundle deals like Buy 3 items and get a discount could get customers to spend more each visit without needing more customers to visit the store.

### 5. Prepare Early for Back to School Season
Both territories make the most money in October every year. Starting promotions on textbooks, laptops, and school supplies in July and August could make that peak season even stronger.
